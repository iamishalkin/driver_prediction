```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
%matplotlib inline
```

```python
for col in df.columns[2:]:
    pyplot.hist(df[(df['target'] == 1)][col],  alpha=0.5, label='1')
    pyplot.hist(df[(df['target'] == 0)][col],  alpha=0.5, label='0')
    pyplot.legend(loc='upper right')
    pyplot.title(col)
    pyplot.show(col)
    sns.countplot(df[(df['target'] == 1)][col])
    plt.figure(col)
```
